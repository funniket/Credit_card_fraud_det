```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load the dataset
data_path = '../Data/card_transdata.csv'
df = pd.read_csv(data_path)
```

```python
# 2. Basic Dataset Information
print("First 5 rows of the dataset:")
print(df.head())

print("\nDataset Information:")
df.info()

print("\nMissing values in each column:")
print(df.isnull().sum())

print("\nSummary Statistics:")
print(df.describe())
```

```python
# 3. Visualizations

# 3.1 Count Plot for the Target Variable
plt.figure()
sns.countplot(x='fraud', data=df)
plt.title("Count of Fraudulent vs Non-Fraudulent Transactions")
plt.xlabel("Fraud (0 = Non-Fraudulent, 1 = Fraudulent)")
plt.ylabel("Count")
plt.show()

```

```python
# 3.2 Histograms with KDE for Numerical Features
numeric_features = ['distance_from_home', 'distance_from_last_transaction', 'ratio_to_median_purchase_price']
for feature in numeric_features:
    plt.figure()
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.show()

```

```python
# 3.3 Boxplots for Numerical Features (to detect outliers)
for feature in numeric_features:
    plt.figure()
    sns.boxplot(x=df[feature])
    plt.title(f"Boxplot of {feature}")
    plt.xlabel(feature)
    plt.show()

```

```python
# 3.4 Count Plots for Binary Features
binary_features = ['repeat_retailer', 'used_chip', 'used_pin_number', 'online_order']
for feature in binary_features:
    plt.figure()
    sns.countplot(x=feature, data=df)
    plt.title(f"Count of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()

```

```python
# 3.6 Correlation Heatmap
plt.figure(figsize=(8, 6))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 3.7 Scatter Plot: Example - Distance from Home vs. distance from last transaction
plt.figure()
sns.scatterplot(data=df, x='distance_from_home', y='distance_from_last_transaction', hue='fraud')
plt.title("Distance from Home vs distance from last transaction")
plt.xlabel("Distance from Home")
plt.ylabel("distance from last transaction")
plt.show()

```
